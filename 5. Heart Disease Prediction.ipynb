```python
#importing dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
heart_disease_data = pd.read_csv('/content/heart_disease_data.csv')
```

1 ---> Heart Defect

0 ---> No Heart Defect

```python
heart_disease_data.groupby('target').mean()
```

```python
#splitting feautures and target
X = heart_disease_data.drop(columns ='target', axis =1)
Y = heart_disease_data['target']
```

```python
print (X)
```

```python
print(Y)
```

splitting into training and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training -- 

Logistic regression mOdel best for binary model

```python
model = LogisticRegression()
```

```python
#train logistic reression model with trianing data
model.fit(X_train,Y_train)
```

Model evaluation -- Accuracy score and metric

```python
#accuracy score of training Data
X_train_prediction =  model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on trainign data is:', training_data_accuracy)
```

```python
#accuracy score of testing Data
X_test_prediction =  model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data is:', test_data_accuracy)
```

Making a Predictive sytem if Heart defect or No hear defect

```python
input_data = (46,1,2,150,231,0,1,147,0,3.6,1,0,2)

#convert input data list to numpyarray
input_data_as_numpyarray = np.asarray(input_data)

#reshape numpy array as we are predicting for 1 instance not all instances
input_data_reshaped = input_data_as_numpyarray.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)


if (prediction[0] == '1'):
  print ('Heart Defect is present')
else:
  print ('Heart Defect is NOT present')
```
